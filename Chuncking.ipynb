```python
import nltk
```

```python
sentence = [("the", "DT"), ("little", "JJ"), ("yellow", "JJ"),("dog", "NN"), ("barked", "VBD"), ("at", "IN"),  ("the", "DT"), ("cat", "NN")]
```

```python
grammer = "NP: {<DT>?<JJ>*<NN>}"
```

```python
cp = nltk.RegexpParser(grammer)
```

```python
result = cp.parse(sentence)
```

```python
print(result)
```

```python
result.draw()
```

```python
text = "I was pleased to know that he applied the concepts in his internship extensively to build a professional tool for Intellect Design Arena. He designed this tool to make the process of configuring a Logical Data Model easier and much faster. It consisted of a User Interface (UI) that can replicate back-end tasks such as inserting data in a database at the click of a button. His task was cut out for him as the tool was being built for J.P. Morgan Chase as a client and hence there was no room for error. I am proud to say that the tool, which he built over a course of two months, was pushed to production at the end of his internship."
words = nltk.word_tokenize(text)
tagged = nltk.pos_tag(words)
```

```python
grammer = "NP: {<RB.?>*<VB.?>*<NNP>+<NN>?}"
cp = nltk.RegexpParser(grammer)
```

```python
chunked = cp.parse(tagged)
print(chunked)
chunked.draw()
```

```python
text = open("lor1.txt")
textdata = text.read()
sentences = nltk.sent_tokenize(textdata)
entities = {}
for i in sentences:
    words = nltk.word_tokenize(i)
    tagged = nltk.pos_tag(words)
    chunks = nltk.ne_chunk(tagged)
    for chunk in chunks:
        if type(chunk) is nltk.Tree:
            t = ''.join(c[0] for c in chunk.leaves())
            entities[t] = chunk.label()
print(entities)
```
