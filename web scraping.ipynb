```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.thehindu.com/news/cities/Tiruchirapalli/iim-tiruchi-makes-progress-in-identifying-permanent-facility-for-chennai-campus/article31916154.ece"
responsepage = requests.get(url)
```

```python
bsoup = BeautifulSoup(responsepage.content, "html.parser")
text = bsoup.get_text()
```

```python
text.replace('\n',"")
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer('\w+')
tokens = tokenizer.tokenize(text)
tokens[:10]
```

```python
import spacy
import en_core_web_sm
nlp = en_core_web_sm.load()
```

```python
doc = nlp(text)
org = []
person=[]
gpe=[]
others = []
for ent in doc.ents:
    if ent.label_ == "ORG":
        org.append(ent.text)
    if ent.label_ == "GPE":
        gpe.append(ent.text)
    if ent.label_ == "PERSON":
        person.append(ent.text)
    else:
        others.append(ent.text)
print("ORG :- ",org)
print("GPE :- ",gpe)
print("PERSON :-", person)
print("OTHERS:- ",others)
```

```python
words  = nlp(text)
```
